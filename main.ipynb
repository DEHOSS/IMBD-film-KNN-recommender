```python
df = pd.DataFrame({ 'movie tittle' : movie_titles_text , 'movie plots' : preprocessed_plots} , columns=['movie tittle' , 'movie plots' ])
```

```python
preprocessed_plots_unnested = []
for i in preprocessed_plots:
    for w in i :
        preprocessed_plots_unnested.append(w)

       
```

```python
counter = Counter(preprocessed_plots_unnested)
dict_counter  = dict(counter)
#list(dict_counter.keys())
dict_counter['young']
```

```python
new_df = pd.DataFrame(columns=[['movie_titels'] + list(dict_counter.keys())])
new_df
```

```python
def idf():
    dict_idf = {}
    for word in list(dict_counter.keys()):
        counter = 0 
        for plot in preprocessed_plots:
            if word in plot :
                counter += 1
        dict_idf[word] = math.log(250/counter) + 1
    return dict_idf
dict_idf = idf()
```

```python
def tf(plot):
    for word in plot:
        tf = plot.count(word)/len(plot)
    return tf
```

```python
plot_tf = []
for plot in preprocessed_plots:
    plot_tf.append(tf(plot))
    
plot_tf
```

```python
rows = []
for i in range(len(preprocessed_plots)):
  movie_rows = []
  for a in list(dict_counter.keys()):
    if a in preprocessed_plots[i] :
      try:
        movie_rows.append(dict_idf[a]*plot_tf[i])
      except:
        pass
    else:
      movie_rows.append(0)
  rows.append(movie_rows)

```

```python
dataframe_dict = {}
for i in range(len(movie_titles_text)):
  dataframe_dict[movie_titles_text[i]] = rows[i]
dataframe_dict
movies_tfidf = pd.DataFrame(dataframe_dict)
movies_tfidf
```

```python
new_plot =  "When the menace known as the Joker wreaks havoc and chaos on the people of Gotham, Batman must accept one of the greatest psychological and physical tests of his ability to fight injustice."
new_preprocessed_plot= [w for w in new_plot.split() if not w.lower() in stop_words]
common_words =[]
for i in new_preprocessed_plot:
    if i in list(dict_counter.keys()):
        common_words.append(i)
common_words
```

```python
new_plot_tf = tf(common_words)
new_plot_rows = []
for a in list(dict_counter.keys()):
    if a in common_words :
      try:
        new_plot_rows.append(dict_idf[a]*new_plot_tf)
      except:
        pass
    else:
      new_plot_rows.append(0)
new_plot_rows
```

```python
dict_dot = {}
for movie in movie_titles_text :
    sum = 0 
    for i in range(len(rows)):
        sum += new_plot_rows[i]*dataframe_dict[movie][i]
    dict_dot[movie] = sum
dict_dot
```
